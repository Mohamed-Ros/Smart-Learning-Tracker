```python
import pandas as pd

df = pd.read_csv("E:\Ai\edu\Smart Learning Tracker\data\processed\preprocessed_data.csv")
```

```python
def evaluate_lesson(row):
    score = row['quiz_score']
    if score < 50:
        return "weak"
    elif score < 80:
        return "medium"
    else:
        return "mastered"

df['lesson_level'] = df.apply(evaluate_lesson, axis=1)

df.to_csv("E:\Ai\edu\Smart Learning Tracker\data\processed\lesson_eval.csv", index=False)
print("Lesson Evaluation saved: data/processed/lesson_eval.csv")
```
