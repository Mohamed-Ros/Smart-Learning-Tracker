```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error
import joblib
import os
```

```python
def train_and_evaluate():
    print("Loading data...")
    data_path = "E:\\Ai\\edu\\Smart Learning Tracker\\data\\processed\\lesson_eval.csv"
    df = pd.read_csv(data_path)
    
    features = ['watched', 'watch_time_min', 'attempts', 'time_spent']
    categorical_features = ['mistake_type']
    target_class = 'lesson_level'
    target_reg = 'quiz_score'
    
    print("Preprocessing features...")
    X = df[features + categorical_features].copy()
    y_class = df[target_class].copy()
    y_reg = df[target_reg].copy()
    
    X = pd.get_dummies(X, columns=categorical_features, drop_first=True)
    
    print("\n--- Training Lesson Level Classifer ---")
    X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X, y_class, test_size=0.2, random_state=42)
    
    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train_c, y_train_c)
    
    y_pred_c = clf.predict(X_test_c)
    print("Classifier Accuracy:", accuracy_score(y_test_c, y_pred_c))
    print("Classification Report:\n", classification_report(y_test_c, y_pred_c))
    
    print("\n--- Training Quiz Score Regressor ---")
    X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X, y_reg, test_size=0.2, random_state=42)
    
    reg = RandomForestRegressor(n_estimators=100, random_state=42)
    reg.fit(X_train_r, y_train_r)
    
    y_pred_r = reg.predict(X_test_r)
    rmse = np.sqrt(mean_squared_error(y_test_r, y_pred_r))
    print(f"Regressor RMSE: {rmse:.2f}")
    
    
    print("\nSaving Models...")
    save_dir = "E:\\Ai\\edu\\Smart Learning Tracker\\models\\saved_models"
    os.makedirs(save_dir, exist_ok=True)
    
    joblib.dump(clf, os.path.join(save_dir, "lesson_level_classifier.pkl"))
    joblib.dump(reg, os.path.join(save_dir, "quiz_score_regressor.pkl"))
    
    joblib.dump(X.columns.tolist(), os.path.join(save_dir, "model_features.pkl"))
    
    print(f"Models and feature configuration saved to {save_dir}")

if __name__ == "__main__":
    train_and_evaluate()
```
