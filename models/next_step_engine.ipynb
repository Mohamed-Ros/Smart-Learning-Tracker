```python
import pandas as pd

df = pd.read_csv("E:\Ai\edu\Smart Learning Tracker\data\processed\lesson_eval.csv")
```

```python
next_steps = []

for idx, row in df.iterrows():
    if row['lesson_level'] == "weak":
        action = "watch_video + practice_easy"
    elif row['lesson_level'] == "medium":
        action = "practice_additional"
    else:
        action = "next_lecture / advanced_content"

    next_steps.append({
        "student_id": row['student_id'],
        "course_id": row['course_id'],
        "lecture_id": row['lecture_id'],
        "lesson_level": row['lesson_level'],
        "next_action": action
    })
```

```python
df_report = pd.DataFrame(next_steps)
df_report.to_csv("E:\Ai\edu\Smart Learning Tracker\data\processed\cumulative_report.csv", index=False)
print("Cumulative Report saved: data/processed/cumulative_report.csv")
```
