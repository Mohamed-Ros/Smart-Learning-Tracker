```python
import pandas as pd
import random
import os
```

```python
os.makedirs("data/raw", exist_ok=True)

students = range(101, 151)
courses = ["Math101", "Physics101", "Chem101"]
lectures = ["Lec1", "Lec2", "Lec3", "Lec4", "Lec5"]
mistakes = ["concept", "calculation", "careless"]
```

```python
data = []
for student in students:
    for course in courses:
        for lecture in lectures:
            watched = random.choice([True, False])
            watch_time = random.randint(5, 60) if watched else 0
            quiz_score = random.randint(30, 100) if watched else 0
            attempts = random.randint(1,3) if watched else 0
            mistake_type = random.choice(mistakes) if watched else None
            time_spent = random.randint(5, 60) if watched else 0
            data.append({
                "student_id": student,
                "course_id": course,
                "lecture_id": lecture,
                "watched": watched,
                "watch_time_min": watch_time,
                "quiz_score": quiz_score,
                "attempts": attempts,
                "mistake_type": mistake_type,
                "time_spent": time_spent
            })
```

```python
df_raw = pd.DataFrame(data)
df_raw.to_csv("data/raw/fake_dataset.csv", index=False)
print("Raw Dataset saved: data/raw/fake_dataset.csv")
```
